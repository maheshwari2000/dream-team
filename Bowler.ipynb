```python
bowl.columns = ['Player',   'Span',    'Mat',   'Inns',  'Balls',   'Mdns',   'Runs',
         'Wkts',    'BBI',    'Ave',   'Econ',     'SR',        '5',       '10',
           'Ct',     'St',   'Team',   'Cost']
bowl['5+'] = bowl['5']+bowl['10']
bowl['Avg Wkts'] = bowl['Wkts']/bowl['Mat']
bowl['Inns%'] = bowl['Inns']/bowl['Mat']
bowl['Inns%'] = bowl['Inns%']*100
bowl = bowl.drop(columns=['Span','Inns','10','5','BBI','Runs','SR','Ct','St'], axis=1)
bowl.head()
```

```python
scoring = {
    'Mat' : 10,
    'Balls' : 10,
    'Mdns' : 5,
    'Wkts' : 25,
    'Ave' : 20,
    'Econ' : 15,
    '5+' : 10,
    'Avg Wkts' : 15,
    'Inns%' : 5,
}
```

```python
bowl.info()
```

```python
for col in scoring.keys():
    score = scoring[col]
    print(col, score)
    bowl[col] = (bowl[col]-bowl[col].min())/(bowl[col].max()-bowl[col].min())
    bowl[col] = bowl[col]*score

bowl.head()
```

```python
bowl['Score'] = bowl[scoring.keys()].sum(axis=1)
bowl.head()
```

```python
from pulp import LpMaximize, LpProblem, LpVariable, lpSum, PULP_CBC_CMD
import pulp
solver = pulp.GLPK_CMD(msg=True)
```

```python
df = bowl.copy()
```

```python
budget = 50
team_size = 11
```

```python
# Define the ILP problem
prob = LpProblem("Team_Selection", LpMaximize)

# Define decision variables
player_vars = LpVariable.dicts("Player", df["Player"], cat="Binary")
player_vars
```

```python
for i in df.index:
    print(df.loc[i, "Player"])
    break
```

```python
# Objective function
prob += lpSum(df.loc[i, "Score"] * player_vars[df.loc[i, "Player"]] for i in df.index)

# Constraints
prob += lpSum(player_vars[df.loc[i, "Player"]] for i in df.index) == team_size
prob += lpSum(df.loc[i, "Cost"] * player_vars[df.loc[i, "Player"]] for i in df.index) <= budget

prob
```

```python
solver = pulp.GLPK_CMD(msg=True)
status = prob.solve(solver)
```

```python
# Solve the problem
prob.solve()
```

```python
# Print the selected team
selected_team = [df.loc[i, "Player"] for i in df.index if player_vars[df.loc[i, "Player"]].value() == 1]
print("Selected Team:", selected_team)
```
