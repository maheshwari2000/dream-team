```python
import pandas as pd
import numpy as np
```

```python
bat = pd.read_excel("Data.xlsx",sheet_name='Sheet1',header=3)

bat['Team'] = bat['Player'].shift(-1)
bat = bat.replace({"-":0})
for col in bat.columns:
    if bat[col].dtype=='O':
        bat[col] = bat[col].str.replace("*","")
        try:
            bat[col] = bat[col].astype('float64')
        except:
            pass

bat = bat[bat['Team'].isin(['Gujarat Titans',
 'Royal Challengers Bengaluru',
 'Mumbai Indians',
 'Delhi Capitals',
 'Lucknow Super Giants',
 'Punjab Kings',
 'Sunrisers Hyderabad',
 'Chennai Super Kings',
 'Rajasthan Royals',
 'Kolkata Knight Riders'])]

bat['Cost'] = np.random.randint(1, 11, size=len(bat))

bat = bat[bat['Team'].isin(['Mumbai Indians','Chennai Super Kings'])]

bat['50+'] = bat[100]+bat[50]
bat['Inns%'] = bat['Inns']/bat['Mat']
bat['Inns%'] = bat['Inns%']*100
bat = bat.drop(columns=['Span','Inns','NO',100,50], axis=1)
bat = bat.fillna(0)

# Can use fantasy point weightage and can normalize it
scoring = {
    'Mat' : 10,
    'Runs' : 25,
    'HS' : 15,
    'Ave' : 20,
    'SR' : 15,
    0 : -5,
    '50+' : 10,
    'Inns%' : 5,
}

for col in scoring.keys():
    score = scoring[col]
    print(col, score)
    bat[col] = (bat[col]-bat[col].min())/(bat[col].max()-bat[col].min())
    bat[col] = bat[col]*score

bat['Score'] = bat[scoring.keys()].sum(axis=1)

bat = bat[['Player','Cost','Score']]
bat['Role'] = 'Batsman'
bat.head()
```

```python
bowl = pd.read_excel("Data.xlsx",sheet_name='Sheet2',header=3)

bowl['Team'] = bowl['Player'].shift(-1)

bowl = bowl.replace({"-":0})

bowl = bowl[bowl['Team'].isin(['Gujarat Titans',
 'Royal Challengers Bengaluru',
 'Mumbai Indians',
 'Delhi Capitals',
 'Lucknow Super Giants',
 'Punjab Kings',
 'Sunrisers Hyderabad',
 'Chennai Super Kings',
 'Rajasthan Royals',
 'Kolkata Knight Riders'])]

bowl['Cost'] = np.random.randint(1, 11, size=len(bowl))

bowl = bowl[bowl['Team'].isin(['Mumbai Indians','Chennai Super Kings'])]

### FIELDING DF
catch = bowl[['Player','Ct','St','Team','Cost']]
catch = catch.fillna(0)

bowl.columns = ['Player',   'Span',    'Mat',   'Inns',  'Balls',   'Mdns',   'Runs',
         'Wkts',    'BBI',    'Ave',   'Econ',     'SR',        '5',       '10',
           'Ct',     'St',   'Team',   'Cost']
bowl['5+'] = bowl['5']+bowl['10']
bowl['Avg Wkts'] = bowl['Wkts']/bowl['Mat']
bowl['Inns%'] = bowl['Inns']/bowl['Mat']
bowl['Inns%'] = bowl['Inns%']*100
bowl = bowl.drop(columns=['Span','Inns','10','5','BBI','Runs','SR','Ct','St'], axis=1)

scoring = {
    'Mat' : 10,
    'Balls' : 10,
    'Mdns' : 5,
    'Wkts' : 25,
    'Ave' : 20,
    'Econ' : 15,
    '5+' : 10,
    'Avg Wkts' : 15,
    'Inns%' : 5,
}

for col in scoring.keys():
    score = scoring[col]
    print(col, score)
    bowl[col] = (bowl[col]-bowl[col].min())/(bowl[col].max()-bowl[col].min())
    bowl[col] = bowl[col]*score

bowl['Score'] = bowl[scoring.keys()].sum(axis=1)

bowl = bowl[['Player','Cost','Score']]
bowl['Role'] = 'Bowler'
bowl.head()
```

```python
catch['Ct'] = (catch['Ct']-catch['Ct'].min())/(catch['Ct'].max()-catch['Ct'].min())
catch['St'] = (catch['St']-catch['St'].min())/(catch['St'].max()-catch['St'].min())
catch['St'] = catch['St'].fillna(0)
catch.head()
```

```python
# Combine batsmen and bowlers into a single DataFrame
players_df = pd.concat([bat, bowl], ignore_index=True)
print(players_df.shape)
players_df = players_df.sort_values(by='Score', ascending=False)
players_df = players_df.drop_duplicates(subset=['Player'])
print(players_df.shape)
players_df.head()
```

```python
from pulp import LpMaximize, LpProblem, LpVariable, lpSum, PULP_CBC_CMD
import pulp
solver = pulp.GLPK_CMD(msg=True)
```

```python
# Budget and team size
budget = 50
team_size = 11
batsman = 6
bowler = 5 

# Define the ILP problem
prob = LpProblem("Team_Selection", LpMaximize)

# Define decision variables
player_vars = LpVariable.dicts("Player", players_df["Player"], cat="Binary")

# Objective function: Maximize total score
prob += lpSum(players_df.loc[i, "Score"] * player_vars[players_df.loc[i, "Player"]] for i in players_df.index)

# Constraints
# 1. Total players = 11
prob += lpSum(player_vars[players_df.loc[i, "Player"]] for i in players_df.index) == team_size

# 2. Budget constraint
prob += lpSum(players_df.loc[i, "Cost"] * player_vars[players_df.loc[i, "Player"]] for i in players_df.index) <= budget

# 3. Team composition constraints
# At Max x batsmen
prob += lpSum(player_vars[players_df.loc[i, "Player"]] for i in players_df.index if players_df.loc[i, "Role"] == "Batsman") >= batsman  #(atleast)

# At least y bowlers
prob += lpSum(player_vars[players_df.loc[i, "Player"]] for i in players_df.index if players_df.loc[i, "Role"] == "Bowler") >= bowler #(atleast)

# Solve the problem
prob.solve()

# Print the selected team
selected_team = [players_df.loc[i, "Player"] for i in players_df.index if player_vars[players_df.loc[i, "Player"]].value() == 1]
print("Selected Team:", selected_team)
```
