```python
bat['50+'] = bat[100]+bat[50]
bat['Inns%'] = bat['Inns']/bat['Mat']
bat['Inns%'] = bat['Inns%']*100
bat = bat.drop(columns=['Span','Inns','NO',100,50], axis=1)
bat = bat.fillna(0)
bat.head()
```

```python
# Can use fantasy point weightage and can normalize it
scoring = {
    'Mat' : 10,
    'Runs' : 25,
    'HS' : 15,
    'Ave' : 20,
    'SR' : 15,
    0 : -5,
    '50+' : 10,
    'Inns%' : 5,
}
```

```python
bat.info()
```

```python
for col in scoring.keys():
    score = scoring[col]
    print(col, score)
    bat[col] = (bat[col]-bat[col].min())/(bat[col].max()-bat[col].min())
    bat[col] = bat[col]*score

bat.head()
```

```python
bat['Score'] = bat[scoring.keys()].sum(axis=1)
bat.head()
```

```python
from pulp import LpMaximize, LpProblem, LpVariable, lpSum, PULP_CBC_CMD
import pulp
solver = pulp.GLPK_CMD(msg=True)
```

```python
df = bat.copy()
```

```python
budget = 50
team_size = 11
```

```python
# Define the ILP problem
prob = LpProblem("Team_Selection", LpMaximize)

# Define decision variables
player_vars = LpVariable.dicts("Player", df["Player"], cat="Binary")
player_vars
```

```python
for i in df.index:
    print(df.loc[i, "Player"])
    break
```

```python
# Objective function
prob += lpSum(df.loc[i, "Score"] * player_vars[df.loc[i, "Player"]] for i in df.index)

# Constraints
prob += lpSum(player_vars[df.loc[i, "Player"]] for i in df.index) == team_size
prob += lpSum(df.loc[i, "Cost"] * player_vars[df.loc[i, "Player"]] for i in df.index) <= budget

prob
```

```python
solver = pulp.GLPK_CMD(msg=True)
status = prob.solve(solver)
```

```python
# Solve the problem
prob.solve()
```

```python
# Print the selected team
selected_team = [df.loc[i, "Player"] for i in df.index if player_vars[df.loc[i, "Player"]].value() == 1]
print("Selected Team:", selected_team)
```
